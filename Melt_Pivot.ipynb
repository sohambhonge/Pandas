```python
import pandas as pd
```

```python
Data = {
    'Name': ['Geet', 'Reet', 'Meet', 'sid', 'rid'],
    'Math': [78, 93, 86, 79, 82],
    'Science': [89, 74, 96, 83, 74],
    'English': [79, 85, 93, 74, 84]
}
df = pd.DataFrame(Data)
```

```python
df
```

```python
melt_df = df.melt(id_vars=["Name"], value_vars=["Math", "Science", "English"], var_name="Subject", value_name="Score") 
```

```python
melt_df
```

```python
df.pivot = melt_df.pivot(
    index="Name",
    columns="Subject",
    values="Score")
```
