### Fill null values

```python
df.isnull() #function shows the true wherever the data contains null value
```

```python
df.isnull().sum() #Function use to show the no. of null values as per columns
```

```python
df.dropna() #Function use to drop the rows that contain any null value
```

```python
df.dropna(axis=1) #Function use to drop the columns that contain any null value
```

```python
df.fillna(0) #Function use to fill null value with 0
```

```python
df["Age"].fillna(df["Age"].mean()) #Function use to fill null values with mean
```

```python
df.ffill() #Function use to fill the null values with previous index value
```

```python
df.bfill() #Function use to fill the null values with next index value
```

### Remove Duplicate Values

```python
df.duplicated() #Function use to find the duplicate rows 
```

```python
df.drop_duplicates() #Function use to remove duplicate rows
```

```python
df.duplicated(subset = ["Name", "Age"]) #Function use to find duplicate values of particular columns
```

```python
df2["Age Group"] = df2["Age"].apply(lambda x: "Adult" if x >= 25 else "Minor")
```

```python
gender_map = {"M": "Male", "F": "Female"}
df2["Gender"] = df2["Gender"].map(gender_map)
```

```python
df2["City"] = df2["City"].replace({"Delhi": "New Delhi", "Mumbai": "New Mumbai"})
```
